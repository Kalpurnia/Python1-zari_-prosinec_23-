```python
import pandas
```

```python
city_temp=pandas.read_csv("temperature.csv")
city_temp.head(40)
```

```python
city_temp.set_index("City", inplace=True)
city_temp.loc["Prague"]
```

```python

hot_cities =city_temp[city_temp["AvgTemperature"]>80]
hot_cities
```

```python

selected_euro_cities = city_temp[(city_temp["AvgTemperature"]>60) & (city_temp["Region"]=="Europe")]
selected_euro_cities
```

```python
extreme_cities = city_temp[(city_temp["AvgTemperature"]>80) | (city_temp["AvgTemperature"]<-20)]
extreme_cities
```

```python
import pytemperature
city_temp["AvgTemperatureCelsia"]=pytemperature.f2c(city_temp["AvgTemperature"])
city_temp
```

```python
hot_cities_Celsia =city_temp[city_temp["AvgTemperatureCelsia"]>30]
hot_cities_Celsia
```

```python
selected_euro_cities_Celsia = city_temp[(city_temp["AvgTemperatureCelsia"]>15) & (city_temp["Region"]=="Europe")]
selected_euro_cities_Celsia
```

```python
extreme_cities_Celsia = city_temp[(city_temp["AvgTemperatureCelsia"]>30) | (city_temp["AvgTemperatureCelsia"]<-10)]
extreme_cities_Celsia
```
