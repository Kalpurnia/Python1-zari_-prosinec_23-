```python
city_temp=pandas.read_csv("temperature.csv")
city_temp.set_index("City", inplace=True)
city_temp_select = city_temp.loc[["Helsinki","Miami Beach", "Tokyo"]]

city_temp_select.reset_index(inplace=True)

city_temp_select = city_temp_select[["City", "AvgTemperature"]]
city_temp_select
```

```python
cities_to_columns = city_temp_select.pivot(columns="City", values="AvgTemperature")# toto přehození nám vysvětloval Jirka v extra lekci v nedeli 

cities_to_columns
```

```python
cities_to_columns.plot(kind="box", whis=[5,95])
plt.ylabel('Průměrná teplota °F) ')
```
